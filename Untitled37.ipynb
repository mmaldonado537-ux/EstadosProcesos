```python
import threading
import time
import random
from collections import deque


# Configuración inicial

tamano_buffer = 5               # máximo de datos que caben
buffer = deque()                # aquí guardamos los datos
espacios_vacios = threading.Semaphore(tamano_buffer)   # cuántos lugares libres hay
datos_disponibles = threading.Semaphore(0)             # cuántos datos hay para consumir
mutex = threading.Semaphore(1)                         # para que no se peleen por el buffer


# Función que hace el sensor (productor)
def sensor(id_sensor):
    while True:
        # Simular que el sensor lee algo (un número aleatorio)
        dato = random.randint(1, 100)
        # Tiempo entre lecturas (para que no sea tan rápido)
        time.sleep(random.uniform(0.5, 2))

        # Esperar a que haya un lugar libre en el buffer
        espacios_vacios.acquire()
        # Bloquear el buffer para que nadie más lo toque mientras metemos el dato
        mutex.acquire()
        buffer.append(dato)
        print(f"Sensor {id_sensor} -> puso {dato} | Buffer: {list(buffer)}")
        mutex.release()
        # Avisar que ya hay un dato nuevo para consumir
        datos_disponibles.release()


def analizador(id_analizador):
    while True:
        # Esperar a que haya algún dato para procesar
        datos_disponibles.acquire()
        # Bloquear el buffer mientras sacamos el dato
        mutex.acquire()
        dato = buffer.popleft()
        print(f"Analizador {id_analizador} -> tomó {dato} | Buffer: {list(buffer)}")
        mutex.release()
        # Avisar que hay un espacio libre en el buffer
        espacios_vacios.release()

        # Simular que tarda un poco en procesar el dato
        time.sleep(random.uniform(0.5, 1.5))



if __name__ == "__main__":
    # Vamos a crear 2 sensores y 2 analizadores
    hilos = []

    for i in range(2):
        t = threading.Thread(target=sensor, args=(i,))
        t.start()
        hilos.append(t)

    for i in range(2):
        t = threading.Thread(target=analizador, args=(i,))
        t.start()
        hilos.append(t)

    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        print("\n¡Simulación detenida! Se acabó el experimento.")
```
